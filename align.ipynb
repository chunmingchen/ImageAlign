```python
# Original implementation with optical flow
base_image = images[-1]


def createCircleMask(height, width, rad):
    # rad: ratio from half of height
    circle = np.zeros((height,width))
    skip_circle = cv2.circle(
      circle,(int(width/2),int(height/2)),
      int(height*.5*rad),1,thickness=-1)
    return circle

def align_image(im, base, max_offset=5, mask_rad=1):
    # Optical flow using Lucas–Kanade method
    # max_offset:
    #   blurring to cover possible offset. Must be odd value, or 0
    # mask_rad:
    #   radius of mask circle over height
    height, width = base_image.shape
    
    if max_offset==0:
        imb = im
        baseb = base
    else:
        imb = cv2.GaussianBlur(im, (max_offset, max_offset), 0)
        baseb = cv2.GaussianBlur(base, (max_offset, max_offset), 0)
    
    dx = cv2.Sobel(baseb,cv2.CV_16S,1,0,ksize=3)/8. # 8: normalize
    dy = cv2.Sobel(baseb,cv2.CV_16S,0,1,ksize=3)/8.
    b = (imb-baseb)
    
    del imb
    del baseb
    
    # mask
    mask = createCircleMask(height, width, mask_rad)
    dx = dx[mask>0].flatten()
    dy = dy[mask>0].flatten()
    b = b[mask>0].flatten()
    
    #Lucas–Kanade method
    A = np.stack([dx, dy]).T
    try:
        invATA = np.linalg.inv(np.dot(A.T, A))
    except:
        print ("Cannot compute inverse matrix! Skip align.")
        return im, [0,0]
    ATb = np.dot(A.T, b)
    v = np.dot(invATA, ATb)*1.5 #why 1.5 gets best alignment?
    
    H = np.float32([[1, 0, v[0]],
                    [0, 1, v[1]]])
    out = cv2.warpAffine(im, H, (width, height))
    return out, v
     
for i in range(len(images)):
    temp = images[i]
    cv2.imshow('image', (temp - base_image+.5))
    for _ in range(0,3): # no more than 3 times
        out, v = align_image(temp, base_image)
        cv2.imshow('aligned', (out - base_image+.5))
        temp = out
        print (i, v)
        k = cv2.waitKey(0)
        if k==27: break
        if np.all(np.abs(v)<1): break
    if k==27:break    
    
print ('Done')                     
```

```python
# Pointwise optical flow
base_image = images[-1]
test_image = images[30]
#base_im = cv2.resize(base_image, None, fx=.25, fy=.25)
#test_im = cv2.resize(test_image, None, fx=.25, fy=.25)
base_im = base_image*255.
test_im = test_image*255.
cv2.imshow('result', base_im)

flow = cv2.calcOpticalFlowFarneback(test_im, base_im,None, 
                                    pyr_scale=0.5, 
                                    levels=3, 
                                    winsize=15, 
                                    iterations=5, 
                                    poly_n=5, 
                                    poly_sigma=1.5, flags=0)
#                                   None, 0.5, 5, 15, 3, 5, 1.5, 0)
#flow = cv2.calcOpticalFlowFarneback(test_im, base_im, None, 0.5, 3, 15, 3, 5, 1.5, 0)
#      flow = cv2.calcOpticalFlowFarneback(prvs,next, None, 0.5, 3, 15, 3, 5, 1.2, 0)


hsv = np.zeros(base_im.shape+(3,)).astype(np.uint8)
hsv[...,1] = 255
mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
hsv[...,0] = ang*180/np.pi/2
hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
bgr = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)

print (np.mean(mag))
cv2.imshow('flow',bgr)
flow.shape
print(np.mean(flow[...,0]))
print(np.mean(flow[...,1]))
```

```python
sz = base_im.shape
im_out = np.zeros(sz, dtype=np.float32)
print (im_out.shape)
for i in range(sz[0]):
    for j in range(sz[1]):
        i1 = int(i+flow[i,j,0])
        j1 = int(j+flow[i,j,1])
        if i1>=0 and j1>=0 and i1<sz[0] and j1<sz[1]:
            im_out[i1][j1] = test_im[i,j]
cv2.imshow('out', np.hstack((im_out, test_im-base_im+128, im_out-base_im+128))/255.)
```

```python
select = np.where((flow[...,0] + flow[...,1]).flatten()>.1)
print(len(select[0]))
ys, xs = np.meshgrid(range(sz[0]), range(sz[1]), indexing='ij')
xs1 = xs+flow[...,0]
ys1 = ys+flow[...,1]
xs = xs.flatten()[select[0]]
ys = ys.flatten()[select[0]]
xs1 = xs1.flatten()[select[0]]
ys1 = ys1.flatten()[select[0]]
M = np.vstack([xs, ys, np.ones(len(xs))]).T
invMTM = np.linalg.inv(M.T @ M)
MTbx = M.T @ xs1
MTby = M.T @ ys1
ax = invMTM @ MTbx
ay = invMTM @ MTby
H = np.vstack([ax, ay])
print (H)
print (np.max(M @ ax - xs1), np.mean(M @ ax - xs1))
print (np.max(M @ ay - ys1), np.mean(M @ ay - ys1))


im_out = cv2.warpAffine(test_im,H,(sz[1],sz[0]))
cv2.imshow('out1', np.hstack((im_out, test_im-base_im+128, im_out-base_im+128))/255.)
```
